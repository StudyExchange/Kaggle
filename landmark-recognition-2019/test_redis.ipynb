## Task

```python
def init_task(task_amount, redis_cli=redis_cli):
    task_arr = [{'id': i, 'topn': 100, 'libary_batch_amount': 20, 'per_process_gpu_memory_fraction': 1./task_amount} for i in range(task_amount)]
    task_ser_arr = [json.dumps(task) for task in task_arr]
    for task_ser in task_ser_arr:
        print(type(task_ser), task_ser)
    redis_cli.delete('task_arr')
    print(redis_cli.lpush('task_arr', *task_ser_arr))
    
init_task(10)
```

```python
def get_one_task(redis_cli=redis_cli):
    task_ser = redis_cli.rpop('task_arr')
    if task_ser:
        return json.loads(task_ser)

for i in range(10):
    task = get_one_task()
    print(type(task), task)
```

## Init test image index

```python
def init_test_idx_arr(iamge_amount, redis_cli=redis_cli):
    test_idx_arr = list(range(iamge_amount))
    for test_idx in test_idx_arr:
        print(type(test_idx), test_idx)
    redis_cli.delete('test_idx_arr')
    print(redis_cli.lpush('test_idx_arr', *test_idx_arr))
    
init_test_idx_arr(10)
```

```python
def get_one_test_image_index(redis_cli=redis_cli):
    test_idx_ser = redis_cli.rpop('test_idx_arr')
    if test_idx_ser:
        return json.loads(test_idx_ser)

for i in range(10):
    test_idx = get_one_test_image_index()
    print(type(test_idx), test_idx)
```
