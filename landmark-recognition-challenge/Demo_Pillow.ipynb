```python
import os

cwd = os.getcwd()
input_folder = os.path.join(cwd, 'input')
img_file = os.path.join(input_folder, 'GreatWall.jpg')
print('input_folder: \t', input_folder)
print('img_file: \t', img_file)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# 打开一个jpg图像文件，注意是当前路径:
im = Image.open(img_file)
plt.imshow(im)
plt.show()

# 获得图像尺寸:
w, h = im.size
print('Original image size: \t%sx%s' % (w, h))
# 缩放到50%:
im.thumbnail((w//2, h//2))
print('Resize image to: \t%sx%s' % (w//2, h//2))
# 把缩放后的图像用jpeg格式保存:
im.save(os.path.join(input_folder, 'thumbnail.jpg'), 'jpeg')

plt.imshow(im)
plt.show()
```

```python
import numpy as np

# Image对象转numpy.array
print('type(im): ', type(im))
np_img = np.array(im)
print(np_img.shape)

# numpy.array转Image对象
im1 = Image.fromarray(np_img)
plt.imshow(im1)
plt.show()

# numpy.array转Image对象，单通道
np_img_2d = np_img[:, :, 1]
print(np_img_2d.shape)
plt.imshow(np_img_2d)
plt.show()

channel = 0
np_img_single = np.zeros(np_img.shape)
np_img_single[:, :, channel] = np_img[:, :, channel]
np_img_single = np_img_single.astype(np.uint8)

# np_img_single = np_img
print(np_img_single.shape)
plt.imshow(np_img_single)
plt.show()

# numpy.array再次转回Image对象
im1 = Image.fromarray(np_img_single)
plt.imshow(im1)
plt.show()
```

```python
def describe(np_im):
    print(np.min(np_im))
    print(np.max(np_im))

describe(np_img)
describe(np_img_single)
```
